```python
fig,ax = plt.subplots(1,2,figsize=(12,4))

sns.countplot(x='year', data = df_day, ax=ax[0], palette=['red','blue'])
ax[0].set_title('Distribution of Years')
ax[0].set_xticks([0,1], ['2011','2012'])
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Count')

sns.barplot(x='year',y='registered', data = df_day, label='Registered Users', ax=ax[1], color='blue')
sns.barplot(x='year',y='casual', data = df_day, label='Casual Users', ax=ax[1], color='red')
ax[1].set_title('Correlation with Registered and Casual Users')
ax[1].set_xticks([0,1], ['2011','2012'])
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Average Users Count')

plt.legend()
fig.show();
```
