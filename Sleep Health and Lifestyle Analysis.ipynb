```python
# Count occurrences of each sleep disorder
disorder_counts = df['Sleep Disorder'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 5))
bars = plt.bar(disorder_counts.index, disorder_counts.values, color=['purple','#4A235A', 'pink'])

# Add title and labels
plt.title('Distribution of Sleep Disorders', fontsize = 16)
plt.xlabel('Sleep Disorder')
plt.ylabel('Count')

# Add values above each bar
for bar, value in zip(bars, disorder_counts.values):
    plt.text(bar.get_x() + bar.get_width() / 2, value, str(value), ha='center', va='bottom')

# Show plot
plt.legend(bars, disorder_counts.index, title='Sleep Disorder', title_fontsize='13', loc='upper right')

plt.show()
```

```python
# Group the DataFrame by 'Sleep Disorder' and 'Gender', and count occurrences
grouped_counts = df.groupby(['Sleep Disorder', 'Gender'])['Gender'].count()

# Plot a pie chart
grouped_counts.plot.pie(autopct='%1.1f%%', figsize=(15, 7), colors=['red', 'pink', 'red', 'pink', 'red', 'pink'])
plt.title('The relationship between Gender and Sleep Disorder', fontsize = 14)
plt.axis('equal')
plt.ylabel('')  # Remove the y-label to avoid repetition
plt.show()
```

```python
# Group the DataFrame by 'Sleep Disorder' and 'Occupation', and count occurrences
occupation_counts = df.groupby(['Sleep Disorder', 'Occupation'])['Occupation'].count()

# Reshape the grouped counts into a pivot table
pivot_table = occupation_counts.unstack()

# Plot a bar plot for each sleep disorder
fig, ax = plt.subplots(figsize=(12, 7))
pivot_table.plot(kind='bar', ax=ax, color=plt.cm.tab20.colors)

# Add title and labels
plt.title('Occupation Distribution by Sleep Disorder', fontsize=16)
plt.xlabel('Sleep Disorder')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.legend(title='Occupation', title_fontsize='13')
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

fig = px.sunburst(df, path=[px.Constant('Sleep quality'), 'Sleep Disorder', 'Quality of Sleep'],
                  color='Sleep Disorder', values='Sleep Duration',
                  color_discrete_sequence=['pink', '#4A235A', '#FFF3FD'],
                  hover_data='Gender')

fig.update_layout(title='The effect of Sleep Quality', font=dict(size=12))

fig.show()
```

```python
# Group the DataFrame by 'Sleep Disorder', 'Quality of Sleep', and aggregate sleep duration
grouped_data = df.groupby(['Sleep Disorder', 'Quality of Sleep'])['Sleep Duration'].mean().unstack()

# Plot a bar graph
fig, ax = plt.subplots(figsize=(12, 8))
grouped_data.plot(kind='bar', ax=ax, color=['pink', '#4A235A', 'green', 'red', 'orange','blue'])

# Add title and labels
plt.title('The effect of Sleep Quality on Sleep', fontsize=20)
plt.xlabel('Sleep Disorder')
plt.ylabel('Average Sleep Duration')

# Add legend
plt.legend(title='Quality of Sleep')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
